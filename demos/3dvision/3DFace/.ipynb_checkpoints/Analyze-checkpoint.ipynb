```julia
using HDF5, JLD, PyPlot, PyCall
@pyimport scipy.misc as scpy
@pyimport scipy.misc as scpy

plt.rc("axes", titlesize="x-large", labelsize="x-large")

RESULTS = Dict()
for folder= ["results/IMAGENET_layer_15_sig10/", "results/IMAGENET_layer_16_sig10/", "results/IMAGENET_layer_17_sig10/", "results/OURS_32_2/" , "results/OURS_64_20/" , "results/OURS_96_120/"]
    list = readdir(folder);
    RESULTS[folder] = Dict()
    for fid = 1:9
        GT_IMAGE = scpy.imread(string("data/",fid,".png"))/255.0
        GT_IMAGE = GT_IMAGE[1:299,1:299,:][:]
        RESULTS[folder][fid] = []
        for j=1:length(list)
            if search(list[j],string("face_",fid,"_")) != 0:-1
#                 println(list[j]," ", fid)
                PXERR = [];
                XX = [];
                allimages = readdir(string(folder, list[j],"/"))
                println(length(allimages))
                for k=0:length(allimages)-3#239
                    image = scpy.imread(string(folder,list[j],"/",k,".png"))/255.0;
                    image = image[1:299,1:299,:][:]
                    t1 = GT_IMAGE - image
                    diff = mse = sum(t1.*t1)/length(GT_IMAGE)
                    PXERR = [PXERR, diff];
                    XX = [XX, k];
#                     println(k)
                end
                RESULTS[folder][fid] = [RESULTS[folder][fid], minimum(PXERR)]
    
#                 if folder == "results/IMAGENET/" 
#                     col = "red"
#                 else
#                     col = "blue"
#                 end
#                 plot(XX, PXERR, color=col, linewidth=3.0)
#                 ylabel("Pixel MSE")
#                 xlabel("Epoch")
#                 plt.savefig("illus.pdf")
            end
        end
        println("face:", fid)
        RESULTS[folder][string("min_",fid)] = minimum(RESULTS[folder][fid])
        println(RESULTS)
    end
end
```

```julia
#analysis

FINAL = Dict()

for folder= ["results/IMAGENET_layer_15_sig10/", "results/IMAGENET_layer_16_sig10/", "results/IMAGENET_layer_17_sig10/", "results/OURS_32_2/" , "results/OURS_64_20/" , "results/OURS_96_120/"]
    FINAL[folder] = []
    for i=1:9
        FINAL[folder] = [FINAL[folder], RESULTS[folder][string("min_",i)]]
    end
    meanv = mean(FINAL[folder])
    stdv = std(FINAL[folder])
    FINAL[folder] = {"mean"=>meanv, "std"=>stdv}
    println("folder:", folder, " " ,FINAL[folder])
end
println()
println(FINAL)

```
